Given two non-negative integers, num1 and num2 represented as string, return the sum of num1 and num2 as a string.

You must solve the problem without using any built-in library for handling large integers (such as BigInteger). You must also not convert the inputs to integers directly.

 

Example 1:

Input: num1 = "11", num2 = "123"
Output: "134"
Example 2:

Input: num1 = "456", num2 = "77"
Output: "533"
Example 3:

Input: num1 = "0", num2 = "0"
Output: "0"
 

Constraints:

1 <= num1.length, num2.length <= 104
num1 and num2 consist of only digits.
num1 and num2 don't have any leading zeros except for the zero itself.

```python
# t O(n + m)
# s O(n + m)
class Solution:
    def addStrings(self, num1: str, num2: str) -> str:
        i, j = len(num1) - 1, len(num2) - 1
        carry = 0
        res = []

        while i >= 0 or j >= 0:
            cur_num1 = int(num1[i]) if i >= 0 else 0
            cur_num2 = int(num2[j]) if j >= 0 else 0
            
            cur_sum = carry + cur_num1 + cur_num2
            carry = cur_sum // 10
            res.append(str(cur_sum % 10))
            i -= 1
            j -= 1
        if carry:
            res.append(str(carry))

        return ''.join(reversed(res))
```

```python
class Solution:
    def addStrings(self, num1: str, num2: str) -> str:
        # Initialize indices for both strings from the end
        i, j = len(num1) - 1, len(num2) - 1
        carry = 0
        result = []
        
        # Iterate while either string still has digits or there is a carry
        while i >= 0 or j >= 0 or carry:
            # Convert character to integer or use 0 if index is out of range
            x1 = int(num1[i]) if i >= 0 else 0
            x2 = int(num2[j]) if j >= 0 else 0
            
            # Calculate the sum of digits and the carry
            value = x1 + x2 + carry
            carry = value // 10  # new carry for the next iteration
            result.append(str(value % 10))  # add the current digit to result
            
            # Move to the next digit
            i -= 1
            j -= 1
        
        # Since we've added digits from right to left, we need to reverse the result
        result.reverse()
        
        # Convert list to string and return
        return ''.join(result)
```
