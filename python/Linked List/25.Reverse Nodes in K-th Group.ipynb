25. Reverse Nodes in k-Group

Given the head of a linked list, reverse the nodes of the list k at a time, and return the modified list.

k is a positive integer and is less than or equal to the length of the linked list. If the number of nodes is not a multiple of k then left-out nodes, in the end, should remain as it is.

You may not alter the values in the list's nodes, only nodes themselves may be changed.

 

Example 1:


Input: head = [1,2,3,4,5], k = 2
Output: [2,1,4,3,5]
Example 2:


Input: head = [1,2,3,4,5], k = 3
Output: [3,2,1,4,5]
 

Constraints:

The number of nodes in the list is n.
1 <= k <= n <= 5000
0 <= Node.val <= 1000
 

Follow-up: Can you solve the problem in O(1) extra memory space?

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def reverseKGroup(self, head: Optional[ListNode], k: int) -> Optional[ListNode]:
        dummy = ListNode(0, head)
        # groupPrev is used to keep track of the node before the current group that is being reversed.
        groupPrev = dummy

        while True:
            '''This loop continues as long as there are groups of k nodes to reverse.
It uses getKth to find the k-th node (kth) from the current position (groupPrev). 
If kth is None, it means there aren't enough nodes left to form a complete group, so the loop breaks.'''
            kth = self.getKth(groupPrev, k)
            if not kth:
                break
            # groupNext points to the node after the k-th node, marking the start of the next group.
            groupNext = kth.next

            # The inner while loop reverses the nodes in the current group. 
            # prev initially points to groupNext, and curr points to the first node in the current group.
            prev, curr = kth.next, groupPrev.next

            while curr != groupNext:
                tmp = curr.next
                curr.next = prev
                prev = curr
                curr = tmp
# After reversing the current group, the previous group's last node (groupPrev.next) should now point to the new first node (kth) of the reversed group.
# groupPrev is updated to the last node of the reversed group (which is the original first node of this group, tmp).
            tmp = groupPrev.next
            groupPrev.next = kth
            groupPrev = tmp
        return dummy.next
    '''This function is used to find the k-th node from the current node. It traverses k nodes starting from the current node. 
    If it reaches the end of the list before traversing k nodes, it returns None. 
    This helps determine if there are enough nodes left to reverse in the current group.'''
    def getKth(self, curr, k):
        while curr and k > 0:
            curr = curr. next
            k -= 1
        return curr 
```
