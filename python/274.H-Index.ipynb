Given an array of integers citations where citations[i] is the number of citations a researcher received for their ith paper, return the researcher's h-index.

According to the definition of h-index on Wikipedia: The h-index is defined as the maximum value of h such that the given researcher has published at least h papers that have each been cited at least h times.

 

Example 1:

Input: citations = [3,0,6,1,5]
Output: 3
Explanation: [3,0,6,1,5] means the researcher has 5 papers in total and each of them had received 3, 0, 6, 1, 5 citations respectively.
Since the researcher has 3 papers with at least 3 citations each and the remaining two with no more than 3 citations each, their h-index is 3.
Example 2:

Input: citations = [1,3,1]
Output: 1
 

Constraints:

n == citations.length
1 <= n <= 5000
0 <= citations[i] <= 1000

```python
# sorting
'''Complexity Analysis
Time complexity : O(nlog⁡n). Comparison sorting dominates the time complexity.
Space complexity : O(1). Most libraries using heap sort which costs O(1) extra space in the worst case.'''
class Solution:
    def hIndex(self, citations: List[int]) -> int:
        # Step 1: Sort the citations in descending order
        citations.sort(reverse=True)
        # Step 2: Iterate through the sorted list and find the H-Index
        # i + 1 represents the number of papers we are considering
        # c is the number of citations for the current paper
        for i, c in enumerate(citations):
            # if the i+1 th paper doesn't have at leat i+1 citations, return i which is the one before i+1
            if c < i + 1:
                return i
        # If we reach here, all papers have at least len(citations) citations
        return len(citations)
'''The h-index is the highest number h such that the researcher has at least h papers with at least h citations each.
so, the index + 1 of each paper in the array should be equal to or less than the number of being citied for the h th paper.
nums[i] >= i + 1
Let's use the array [3, 0, 6, 1, 5]:
Sort the citations array in descending order: [6, 5, 3, 1, 0]
Create a chart:
Paper 1: 6 citations
Paper 2: 5 citations
Paper 3: 3 citations
Paper 4: 1 citation
Paper 5: 0 citations

Determine the H-Index:
The h-index is the highest number h such that the researcher has at least h papers with at least h citations each.
From the sorted array, we can see:
Paper 1: 6 citations (at least 1 paper with at least 1 citation)
Paper 2: 5 citations (at least 2 papers with at least 2 citations)
Paper 3: 3 citations (at least 3 papers with at least 3 citations)
Paper 4: 1 citation (at least 4 papers with at least 4 citations : not satisfied)
Paper 5: 0 citations (at least 5 papers with at least 5 citations : not satisfied)
The highest value h that satisfies the condition is 3.
Given array: [10, 9, 5, 3, 3, 2, 1], 
paper 4th (index 3): nums[3] < 4, which means paper 4 doesnt have 4 citations (it has only 3 citations), so h stops at 3, the largest is h at 3.
At least 3 papers have been cited at least 3 times.
Check the condition for each index:
For each position 
i, we check if the number of citations for that paper (citations[i]) is greater than or equal to i+1.
Iteration
Let's iterate through the sorted array and see when the condition fails:
i=0: citations[0]=10, condition: 10≥1 (True)
i=1: citations[1]=9, condition: 9≥2 (True)
i=2: citations[2]=5, condition: 5≥3 (True)
i=3: citations[3]=3, condition: 3≥4 (False)
i=4: citations[4]=3, condition: 3≥5 (False)
i=5: citations[5]=2, condition: 2≥6 (False)
i=6: citations[6]=1, condition: 1≥7 (False)
Why the h-index is 3:
The h-index is the highest h such that there are at least h papers with h or more citations.
At i=2, the condition is satisfied (5≥3).
At i=3, the condition fails (3≥4).
Thus, the h-index is 3 because there are 3 papers with at least 3 citations each.
'''
```
