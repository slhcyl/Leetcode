Implement the BSTIterator class that represents an iterator over the in-order traversal of a binary search tree (BST):

BSTIterator(TreeNode root) Initializes an object of the BSTIterator class. The root of the BST is given as part of the constructor. The pointer should be initialized to a non-existent number smaller than any element in the BST.
boolean hasNext() Returns true if there exists a number in the traversal to the right of the pointer, otherwise returns false.
int next() Moves the pointer to the right, then returns the number at the pointer.
Notice that by initializing the pointer to a non-existent smallest number, the first call to next() will return the smallest element in the BST.

You may assume that next() calls will always be valid. That is, there will be at least a next number in the in-order traversal when next() is called.

 

Example 1:


Input

[[[7, 3, 15, null, null, 9, 20]], [], [], [], [], [], [], [], [], []]
Output
[null, 3, 7, true, 9, true, 15, true, 20, false]

Explanation
BSTIterator bSTIterator = new BSTIterator([7, 3, 15, null, null, 9, 20]);
bSTIterator.next();    // return 3
bSTIterator.next();    // return 7
bSTIterator.hasNext(); // return True
bSTIterator.next();    // return 9
bSTIterator.hasNext(); // return True
bSTIterator.next();    // return 15
bSTIterator.hasNext(); // return True
bSTIterator.next();    // return 20
bSTIterator.hasNext(); // return False

The `BSTIterator` class is designed to iterate over a binary search tree (BST) in an in-order fashion. Let's analyze the time and space complexity of this solution.

### Time Complexity

1. **Constructor (`__init__` method)**:
   - The constructor initializes the stack by traversing the leftmost path of the tree.
   - In the worst case, this traversal takes \( O(h) \) time, where \( h \) is the height of the tree. For a balanced BST, \( h \) is \( O(\log n) \), and for a skewed tree, \( h \) is \( O(n) \).

2. **`next` method**:
   - The `next` method does two main things: it pops an element from the stack and then pushes all the left children of the right child of the popped node onto the stack.
   - Each node is pushed and popped from the stack exactly once.
   - Hence, the amortized time complexity for the `next` method is \( O(1) \). This is because each node is visited only once during the traversal and each push/pop operation on the stack is \( O(1) \).

3. **`hasNext` method**:
   - The `hasNext` method simply checks if the stack is non-empty, which is an \( O(1) \) operation.

### Space Complexity

- The space complexity is determined by the size of the stack.
- In the worst case, the stack can hold all the nodes from the root to the leftmost leaf, which means the stack can contain \( O(h) \) nodes, where \( h \) is the height of the tree.
- For a balanced BST, the height \( h \) is \( O(\log n) \), leading to a space complexity of \( O(\log n) \).
- For a skewed tree, the height \( h \) is \( O(n) \), leading to a space complexity of \( O(n) \).

### Summary

- **Time Complexity**:
  - Constructor: \( O(h) \)
  - `next`: Amortized \( O(1) \)
  - `hasNext`: \( O(1) \)

- **Space Complexity**:
  - \( O(h) \), where \( h \) is the height of the tree. This is \( O(\log n) \) for a balanced tree and \( O(n) \) for a skewed tree.

```python
# t hasNext() O(1), and next() takes O(1) but
# s O(h)
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class BSTIterator:

    def __init__(self, root: Optional[TreeNode]):
        self.stack = []
        while root:
            self.stack.append(root)
            root = root.left

    def next(self) -> int:
        res = self.stack.pop()
        cur = res.right
        while cur:
            self.stack.append(cur)
            cur = cur.left
        return res.val

    def hasNext(self) -> bool:
        return self.stack != []
        


# Your BSTIterator object will be instantiated and called as such:
# obj = BSTIterator(root)
# param_1 = obj.next()
# param_2 = obj.hasNext()
```
