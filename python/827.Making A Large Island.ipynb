You are given an n x n binary matrix grid. You are allowed to change at most one 0 to be 1.

Return the size of the largest island in grid after applying this operation.

An island is a 4-directionally connected group of 1s.

 

Example 1:

Input: grid = [[1,0],[0,1]]
Output: 3
Explanation: Change one 0 to 1 and connect two 1s, then we get an island with area = 3.
Example 2:

Input: grid = [[1,1],[1,0]]
Output: 4
Explanation: Change the 0 to 1 and make the island bigger, only one island with area = 4.
Example 3:

Input: grid = [[1,1],[1,1]]
Output: 4
Explanation: Can't change any 0 to 1, only one island with area = 4.
 

Constraints:

n == grid.length
n == grid[i].length
1 <= n <= 500
grid[i][j] is either 0 or 1.

```python
# T O(N^2)
# S O(N^2)
class Solution:
    def largestIsland(self, grid: List[List[int]]) -> int:
        n = len(grid)    

        def dfs(r, c, index):
            if r < 0 or c < 0 or r >= n or c >= n or grid[r][c] != 1:
                return 0
            grid[r][c] = index 
            return 1 + dfs(r + 1, c, index) + dfs(r - 1, c, index) + dfs(r, c + 1, index) + dfs(r, c - 1, index)

        island_size = {0:0}    
        index = 2
        for r in range(n):
            for c in range(n):
                if grid[r][c] == 1:
                    island_size[index] = dfs(r, c, index)
                    index += 1
                    
        max_size = max(island_size.values())
        for r in range(n):
            for c in range(n):
                if grid[r][c] == 0:
                    seen = set()
                    for dr, dc in [(1, 0), (-1, 0), (0, 1), (0, -1)]:
                        nr, nc = dr + r, dc + c
                        if 0 <= nr < n and 0 <= nc < n:
                            seen.add(grid[nr][nc])
                    max_size = max(max_size, sum(island_size[index] for index in seen) + 1)

        return max_size
```

```python
class Solution:
    def largestIsland(self, grid: List[List[int]]) -> int:
        n = len(grid)

        def dfs(r, c, index):
            # Base case for DFS: out of bounds, already visited, or water
            if r < 0 or c < 0 or r >= n or c >= n or grid[r][c] != 1:
                return 0
            # Mark the current cell with the island index
            grid[r][c] = index
            # Count the current cell and explore all 4 directions
            return 1 + dfs(r + 1, c, index) + dfs(r - 1, c, index) + dfs(r, c + 1, index) + dfs(r, c - 1, index)
            # print(f'step3: r is {r},c is {c}, index is {index}')

        # Step 1: Identify all islands and their sizes
        island_size = {0: 0}  # Dictionary to hold island sizes, with 0 initialized
        index = 2  # Start island indexing from 2 since 0 and 1 are used in the grid
        for r in range(n):
            for c in range(n):
                if grid[r][c] == 1:
                    island_size[index] = dfs(r, c, index)
                    # print(f'step1: r is {r},c is {c}, index is {index}')
                    index += 1

        # Step 2 & 3: Calculate the size of the potential island for each zero
        max_size = max(island_size.values())  # Handle the case where no zero is flipped
        for r in range(n):
            for c in range(n):
                if grid[r][c] == 0:
                    seen = set()  # To hold unique island indexes around the zero
                    for dr, dc in [(1, 0), (-1, 0), (0, 1), (0, -1)]:
                        nr, nc = r + dr, c + dc
                        if 0 <= nr < n and 0 <= nc < n:
                            seen.add(grid[nr][nc])
                        # print(f'step2: dr is {dr},dc is {dc}, seen is {seen}')
                    # Calculate potential island size by summing unique adjacent island sizes and 1 for the current cell
                    max_size = max(max_size, sum(island_size[index] for index in seen) + 1)

        return max_size
```
